```python
df['Ship Date'] = pd.to_datetime(df['Ship Date'], errors='coerce')
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
```

```python
check_list = df.select_dtypes(include='object').columns.tolist()

def unique_value_check(df,col_name):
    '''Unique values check'''
    x = df[col_name].value_counts(dropna= False)
    print(x)

for col_name in check_list :
    print(col_name)
    unique_value_check(df,col_name)
    print('\n')
```

```python
df[df['Order ID']=="CA-2017-100111"]
```

```python
df['Days to Ship'] = (df['Ship Date'] - df['Order Date']).dt.days
```

```python
df['Profit Margin %'] = df['Profit']/df['Sales']
```

```python
df['Loss Indicator'] = np.where(df['Profit']<0,1,0)
```
